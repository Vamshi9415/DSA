```python
# Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

# You may assume that each input would have exactly one solution, and you may not use the same element twice.

# You can return the answer in any order.

 

# Example 1:

# Input: nums = [2,7,11,15], target = 9
# Output: [0,1]
# Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
# Example 2:

# Input: nums = [3,2,4], target = 6
# Output: [1,2]
# Example 3:

# Input: nums = [3,3], target = 6
# Output: [0,1]
def twoSum(nums, target) :

        hashmap = {}

        result = []
        for i in range(len(nums)):
            if target - nums[i] in hashmap :
                result.append(i)
                result.append(hashmap[target-nums[i]])
                break
            else:
                hashmap[nums[i]] = i
        
        return result
```

Given an array of integers arr[]  and a number k. Return the maximum sum of a subarray of size k.

```python
def maxSubarraySum(arr, k):
        n = len(arr)
        
        # first calculate the sum of the first k elements and then add new and subtract last k element
        sum = 0
        for i in range(k):
            sum += arr[i]
        max_sum = sum
        
        # or sum = sum(arr[:k])
        
        
        for i in range(k,n):
            sum = sum -arr[i-k] +arr[i]
            max_sum = max(max_sum,sum)
        
        return max_sum
    
```

Maximum Sum Subarray (Kadane's algorithm)

Find the contiguous subarray with the largest sum.

```python
# Input: arr = [1, -2, 3, 4, -1, 2, 1, -5, 4]
# Output: 9
# (Subarray: [3, 4, -1, 2, 1])
# Sliding window works when:
# Numbers are non-negative
# Window size is fixed

# Kadane handles:
# Negative numbers
# Variable-length subarrays
# Unknown best window size
def kadane(arr):
    n = len(arr)
    
    current_sum = arr[0]
    max_sum = arr[0]
    
    for i in range(1,n):
        current_sum = max(arr[i],current_sum+arr[i])
        max_sum = max(current_sum,max_sum)
    
    return max_sum

kadane([1, -2, 3, 4, -1, 2, 1, -5, 4])
```

Minimum Sum Subarray

```python
# Input
# arr = [3, -4, 2, -3, -1, 7, -5]


# Output
# Minimum Sum: -6
# Subarray: [-4, 2, -3, -1]

def min_sum_subarray(arr):
    n = len(arr)
    
    min_sum = arr[0]
    curr_sum = arr[0]
    
    for i in range(1,n):
        curr_sum = min(curr_sum+arr[i],arr[i])
        min_sum = min(curr_sum,min_sum)
    
    return min_sum

min_sum_subarray([3, -4, 2, -3, -1, 7, -5])
```

Maximum Sum Circular Subarray

Question: 
Given a circular integer array, find the maximum possible sum of a contiguous subarray.
(The subarray may wrap around the end of the array.)

```python
# Example 1
# Input
# arr = [8, -1, 3, 4]

# Output
# Maximum Circular Sum: 15

# In a normal array, subarray must be contiguous and stay inside bounds.

# In a circular array:
# The last element connects to the first
# Subarray can wrap once from end → start
# Each element can be used at most once

# Example:

# [8, -1, 3, 4]

# Valid circular subarray:
# [3, 4, 8]  → wraps

# Invalid:
# [8, -1, 3, 4, 8]  → repeats elements 

# Any circular max subarray looks like this:

# [ 8 | -1 | 3  4 ]
#        ↑
#    minimum sum part


#  Instead of finding the wrap subarray
#  Exclude the worst (minimum sum) subarray

# Max Circular Sum = Total Array Sum − Minimum Subarray Sum

def max_sum_circular_subarray(a):
    total_sum = a[0]
    curr_max = max_sum = a[0]
    curr_min = min_sum = a[0]

    for i in range(1, len(a)):
        x = a[i]
        total_sum += x

        curr_max = max(x, curr_max + x)
        max_sum = max(max_sum, curr_max)

        curr_min = min(x, curr_min + x)
        min_sum = min(min_sum, curr_min)

    if max_sum < 0:
        return max_sum

    result = max(max_sum, total_sum - min_sum)
    return result

max_sum_circular_subarray([8, -1, 3, 4])

```

Fruits in Basket

```python
# You are visiting a farm that has a single row of fruit trees arranged from left to right. The trees are represented by an integer array fruits where fruits[i] is the type of fruit the ith tree produces.

'''
Problem:
You want to collect as much fruit as possible. However, the owner has some strict rules:
- You only have two baskets, and each basket can only hold a single type of fruit. There is no limit on the amount of fruit each basket can hold.
- Starting from any tree of your choice, you must pick exactly one fruit from every tree (including the start tree) while moving to the right. The picked fruits must fit in one of your baskets.
- Once you reach a tree with fruit that cannot fit in your baskets, you must stop.
Given the integer array fruits, return the maximum number of fruits you can pick.
'''

# Approach: Sliding Window + Hashmap
from collections import defaultdict

def totalFruit(fruits):
    '''
    Returns the maximum number of fruits that can be picked with at most two types of fruits in baskets.
    Args:
        fruits (List[int]): List of integers representing fruit types on each tree.
    Returns:
        int: Maximum number of fruits that can be picked.
    '''
    freq = defaultdict(int)  # Dictionary to count fruit types in the current window
    max_fruits = 0           # Store the maximum window size found
    left = 0                 # Left pointer of the sliding window

    for right in range(len(fruits)):
        freq[fruits[right]] += 1  # Add current fruit to the window

        # If more than 2 types, shrink window from the left
        while len(freq) > 2:
            freq[fruits[left]] -= 1
            if freq[fruits[left]] == 0:
                del freq[fruits[left]]
            left += 1

        # Update max_fruits if current window is larger
        max_fruits = max(max_fruits, right - left + 1)

    return max_fruits

# ------------------- Test Cases -------------------
if __name__ == "__main__":
    # Test case 1: All same type
    print(totalFruit([1,1,1,1]))  # Output: 4
    # Test case 2: Example from problem statement
    print(totalFruit([1,2,1]))    # Output: 3
    print(totalFruit([0,1,2,2]))  # Output: 3
    print(totalFruit([1,2,3,2,2])) # Output: 4
    # Test case 3: Only one type
    print(totalFruit([5]))        # Output: 1
    # Test case 4: More than two types in a row
    print(totalFruit([1,2,3,4,5])) # Output: 2
    # Test case 5: Longest window in the middle
    print(totalFruit([1,2,1,2,3,2,2,1])) # Output: 5
```
