```python
def print13(n):
    for i in range(1,n+1):
        for j in range(1,i+1):
            print(j,end = " ")
        for k in range(1,2*n - 2*i +1):
            print("-",end = " ")
        for l in range(1,i+1):
            print(l,end = " ")
        print()

print13(4)
```

```python
def print16(n):
    for i in range(n):
        for j in range(ord('A'),ord('A')+n-i):
            print(chr(j),end = " ")
        print()
print16(5)
```

```python
def print18(n):
    for i in range(n):
        # spaces
        for j in range(n - i - 1):
            print("-", end=" ")
        
        # characters
        char = 'A'
        mid = (2 * i) // 2

        for k in range(2 * i + 1):
            if k <= mid:
                print(char, end=" ")
                char = chr(ord(char) + 1)
            else:
                char = chr(ord(char) - 2)
                print(char, end=" ")
                char = chr(ord(char) + 1)

        # spaces
        for l in range(n - i - 1):
            print("-", end=" ")
        
        print()

print18(4)
```

```python
def print19(n):
    for i in range(n):
        start_char = chr(ord('A') + n - 1 - i)   # E, D, C, B, A

        for j in range(i + 1):
            print(chr(ord(start_char) + j), end=" ")
        print()

print19(5)
```

```python
def print20_1(n):
    for i in range(n):
        for j in range(n-i):
            print("*",end = " ")
        for k in range(2*i) :
            print("-", end = " ")
        for l in range(n-i):
            print("*",end = " ")
        print()

print20_1(5)

def print20_2(n):
    for i in range(n):
        for j in range(i+1):
            print("*",end = " ")
        for k in range(2*n-(2*i)-2):
            print("-",end = " ")
        for l in range(i+1):
            print("*",end = " ")
        print()

print20_2(5)
            
```
