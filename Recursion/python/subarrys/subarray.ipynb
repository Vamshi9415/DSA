```python
# 1. The Problem (Without Prefix Sum)
# Imagine you have an array of daily expenses for a week: A = [10, 20, 10, 5, 15]

# If I ask you: "What is the total expense from Day 1 to Day 3 (indices 1 to 3)?"

# The Slow Way: You have to loop through the array: 20 + 10 + 5 = 35.

# The Issue: If the array has 1 million items and I ask you 1 million questions about different ranges, your computer will freeze because it has to re-calculate the loops every single time.

# 2. The Solution (Prefix Sum Array)
# Instead of calculating on the fly, we pre-calculate a "Running Total" array P.

# P[i] contains the sum of all numbers in the original array from index 0 up to index i.

# Let's build it: Original A: [10, 20, 10, 5, 15]

# Index 0: 10

# Index 1: 10 + 20 = 30

# Index 2: 10 + 20 + 10 = 40

# Index 3: 10 + 20 + 10 + 5 = 45

# Index 4: 10 + 20 + 10 + 5 + 15 = 60

# Prefix P: [10, 30, 40, 45, 60]

# Sum(L, R) = P[R] - P[L-1]

# Goal: Find sum of A from index 1 to 3 (20, 10, 5). We know the answer is 35.
# Using Prefix Array:L = 1, R = P[3] - P[1-1] -> P[3] - P[0]
# Values: P[3] is 45.
# P[0] is 10.
# Calculation: 45 - 10 = 35.


def prefixsum(arr,L,R):
    n = len(arr)
    
    p = [0]*(n+1)
    
    for i in range(n):
        p[i+1] = p[i]+arr[i]

   # index L to R
    result = p[R+1] -p[L]
    
    print(result)

prefixsum([1,2,3,4],0,2)
```

```python
from collections import defaultdict
arr = [1,-1,2,-2]

n = len(arr)

p = [0]*(n+1)

for i in range(n):
    p[i+1] = p[i] + arr[i]

sum_map = defaultdict(list)

for start in range(n):
    for end in range(start,n):
        current_sum = p[end+1] - p[start]
        sum_map[current_sum].append((start,end))

print(sum_map)

# defaultdict(<class 'list'>, {1: [(0, 0), (1, 2)], 0: [(0, 1), (0, 3), (2, 3)], 2: [(0, 2), (2, 2)], -1: [(1, 1), (1, 3)], -2: [(3, 3)]})

#  i need to find the non overlapping pairs (l1, r1) and (l2, r2) are non-overlapping if r1 < l2
total = 0
for intervals in sum_map.values():
    intervals.sort(key = lambda x : (x[1],x[0]) )
    print(intervals)
    k = len(intervals)
    for i in range(k):
        l2,r2 = intervals[i]
        for j in range(i):
            l1,r1 = intervals[j]
            if r1<l2 :
                total +=1
        
print(total)
    
    
```
