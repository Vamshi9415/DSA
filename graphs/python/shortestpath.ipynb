Dijkstra's Algo

```python
import heapq

def dijkstra(n, edges, src):
    graph = [[] for _ in range(n)]
    for u, v, w in edges:
        graph[u].append((v, w))
        graph[v].append((u, w))  
    inf = float('inf')
    distance = [inf] * n
    distance[src] = 0

    pq = [(0, src)]  

    while pq:
        d, node = heapq.heappop(pq)

        if d > distance[node]:
            continue

        for neigh, wt in graph[node]:
            new_dist = d + wt
            if new_dist < distance[neigh]:
                distance[neigh] = new_dist
                heapq.heappush(pq, (new_dist, neigh))

    return distance


if __name__ == '__main__':
    n = 5
    edges = [
        (0, 1, 2),
        (0, 2, 4),
        (1, 2, 1),
        (1, 3, 7),
        (2, 4, 3),
        (3, 4, 1)
    ]
    src = 0

    result = dijkstra(n, edges, src)
    print('Shortest distances from source', src, ':', result)
```

Printing Path

```python
import heapq

def printing_dij_path(n, graph, src, dest):
    inf = float('inf')
    dist = [inf] * n
    parent = [-1] * n
    dist[src] = 0

    pq = [(0, src)]

    while pq:
        d, node = heapq.heappop(pq)

        if d > dist[node]:
            continue

        for neigh, wt in graph[node]:
            newdist = d + wt

            if newdist < dist[neigh]:
                dist[neigh] = newdist
                parent[neigh] = node
                heapq.heappush(pq, (newdist, neigh))

    if dist[dest] == inf:
        print(f'No path from {src} to {dest}')
        return []

    path = []
    cur = dest
    while cur != -1:
        path.append(cur)
        cur = parent[cur]
    path.reverse()

    print('distance array ',dist)
    print('Parent array:', parent)
    print(f'Shortest distance from {src} to {dest}:', dist[dest])
    print('Path:', path)
    return path


if __name__ == '__main__':
    n = 6
    edges = [
        (0, 1, 4),
        (0, 2, 1),
        (2, 1, 2),
        (1, 3, 1),
        (2, 3, 5),
        (3, 4, 3),
        (4, 5, 1),
    ]

    graph = [[] for _ in range(n)]
    for u, v, w in edges:
        graph[u].append((v, w))
        graph[v].append((u, w))

    src, dest = 0, 5
    printing_dij_path(n, graph, src, dest)
```

```python
import heapq
# n = 4
# graph = {
#     0: [(2,1),(6,2)],
#     1: [(2,0),(3,2),(8,3)],
#     2: [(6,0),(3,1),(7,3)],
#     3: [(8,1),(7,2)]
# }

n = int(input())
graph = eval(input())

def prims(n,graph):
    visited = [False]*n
    
    pq = [(0,0)]
    total = 0
    while pq :
        d,node = heapq.heappop(pq)
        
        if visited[node] :
            continue 
        total = total+d
        visited[node] = True 
        
        for wt,neigh in graph[node]:
            if not visited[neigh]:
                heapq.heappush(pq,(wt,neigh))
    
    return total

prims(n,graph)
```
