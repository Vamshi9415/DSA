TRAVERSAL OF CONNECTED GRAPH BFS

```python
from collections import deque
def bfs_connected_graph(v,edges):
    
    adj = [ [] for _ in range(v)]

    for u,node_v in edges :
        adj[u].append(node_v)
        adj[node_v].append(u)
    
    q = deque()
    visited = [False]*v
    visited[0] =True
    q.append(0)
    
    bfs = []
    
    while q :
        node = q.popleft()
        
        bfs.append(node)

        for neigh in adj[node]:
            if not visited[neigh]:
                visited[neigh] = True
                q.append(neigh)
    
    return bfs

v = 5

edges = [
(0, 1),
(0, 2),
(1, 3),
(2, 4)
]

bfs_connected_graph(v,edges)


                
    
    
```

PROBLEM 2: BFS Traversal of a Disconnected Graph


Question:
Given a graph that may contain disconnected components, perform BFS on all components and return the final list.


Example:


Input:
V = 6  
Edges:
0 1  
2 3  
4 5  


Output:
0 1 2 3 4 5


```python
from collections import deque

def bfs_disconnected_graph(v,edges):
    adj = [[] for _ in range(v)]
    
    for u,node_v in edges:
        adj[u].append(node_v)
        adj[node_v].append(u)
    
    
    visited = [False]*v
    bfs = []
    
    for i in range(v):
        if not visited[i]:
            visited[i] = True
            
            q = deque()
            
            q.append(i)
            
            while q :
                node = q.popleft()
                
                bfs.append(node)
                
                for neigh in adj[node]:
                    if not visited[neigh]:
                        visited[neigh] = True
                        
                        q.append(neigh)
    
    return bfs

v = 6

edges = [(0,1),(2,3),(4,5)]

bfs_disconnected_graph(v,edges)
            
```

PROBLEM 3: Check if a Graph is Bipartite using BFS


Question:
Given an undirected graph, check whether it is bipartite using BFS.


Example Input:


0 - 1  
|       |  
3 - 2



Output:

Graph is Bipartite

```python
# A graph is bipartite if:

# You can divide all vertices into two sets: Set A and Set B

# No two adjacent nodes should be in the same set.

# Think of it as 2-coloring the graph (color with 0 or 1).
# If we can successfully color the entire graph → Bipartite
# If at some point an adjacent node has same color → Not Bipartite

def is_bipartite(v,adj):
    color = [-1]*v
    
    color[0] = 1
    
    q = deque()
    q.append(0)
    visited = [False]*v
    visited[0] =True
    while q :
        node  = q.popleft()
        
        for neigh in adj[node]:
            if not visited[neigh]:
                visited[neigh] = True
                color[neigh] = 1- color[node]
            
            elif visited[neigh] and color[neigh] == color[node]:
                return False
    
    return True

adj = [
    [1, 3],  # 0
    [0, 3],  # 1
    [1, 3],  # 2
    [0, 2]   # 3
]

is_bipartite(4,adj)
                
        
```

 Number of Connected Components


Question:
Using BFS, count how many connected components exist in the graph.


Example:


Input:
V = 7

Edges:

0 1

2 3

4 5

\# 6 is isolated


Output:
4 connected components

```python
def numberofconnectedcomp(v,edges):
    
    visited = [False]*v
    
    count = 0 
    
    adj = [[] for _ in range(v)]
    
    for u,node_v in edges:
        adj[u].append(node_v)
        adj[node_v].append(u)
    
    for i in range(v):
        if not visited[i]:
            visited[i] = True
            count +=1
            q = deque()
            q.append(i)
            while q :
                node = q.popleft()
                
                for neigh in adj[node]:
                    
                    if not visited[neigh] :
                        visited[neigh] =True
                        q.append(neigh)
    
    return count 
v = 7
edges = [(0,1), (2,3), (4,5)] # 6 is isolated
numberofconnectedcomp(v,edges)
                    
                
                
```

Detect cycle in undirected graph

```python
# 0 - 1 - 2
# |       |
# 3 ------ 
# Cycle Exists

def is_cycle_in_undirected_graph(v,edges):
    adj = [[] for _ in range(v)]
    for u, node_v in edges:
        adj[u].append(node_v)
        adj[node_v].append(u)
    
    
    visited = [False]*v

    visited[0] =True
    q = deque()
    
    q.append((0,-1))
    
    while q :
        node, parent = q.popleft()
        
        for neigh in adj[node]:
            if not visited[neigh] :
                visited[neigh] = True
                
                q.append((neigh,node))

            elif visited[neigh]  == True and neigh != parent :
                return True
    
    return False

V0 = 4
edges0 = [(0,1),(1,2),(2,3),(3,0)]


print(is_cycle_in_undirected_graph(V0, edges0))
            
                
    
    
```
