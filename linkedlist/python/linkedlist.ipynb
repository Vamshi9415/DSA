palindrome of a linked list

 Given the head of a singly linked list, return true if it is a palindrome or false otherwise.


Assume linked list:

```
head → 1 → 2 → 3 → None
```

Initially:

```
prev = None
current = 1
```

---

##  Iteration 1

### Before changes:

```
prev = None
current = 1 → 2 → 3 → None
```

### Step 1: save next

```python
next_node = current.next
```

So:

```
next_node = 2
```

### Step 2: reverse link

```python
current.next = prev
```

So `1.next = None`

Now list looks like:

```
1 → None
```

### Step 3: move prev forward

```python
prev = current
```

So:

```
prev = 1 → None
```

### Step 4: move current forward

```python
current = next_node
```

So:

```
current = 2 → 3 → None
```

 After iteration 1:

```
prev → 1 → None
current → 2 → 3 → None
```

---

##  Iteration 2

### current = 2

Step 1:

```
next_node = 3
```

Step 2:

```
2.next = prev
2.next = 1
```

So now:

```
2 → 1 → None
```

Step 3:

```
prev = 2
```

Step 4:

```
current = 3
```

 After iteration 2:

```
prev → 2 → 1 → None
current → 3 → None
```

---

##  Iteration 3

### current = 3

Step 1:

```
next_node = None
```

Step 2:

```
3.next = prev
3.next = 2
```

So:

```
3 → 2 → 1 → None
```

Step 3:

```
prev = 3
```

Step 4:

```
current = None
```

Loop ends.

 Final:

```
prev → 3 → 2 → 1 → None
```

---


````python
# Input: head = [1,2,2,1]
# Output: true

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverse(head):
    if head is None:
        return None
    
    prev = None
    current = head
    # head->1->2->3->None
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    
    return prev

def is_palindrome(head):
    if head is None:
        return True  # Fixed: empty list is palindrome
    slow = head
    fast = head
    
    # Fixed: correct condition to find the middle
    while fast.next and fast.next.next:
        slow = slow.next
        fast = fast.next.next

    prev = None
    current = slow.next  
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    
    right = prev
    
    left = head
    
    while right:
        if left.val != right.val:
            return False
        left = left.next
        right = right.next
    
    return True

def build_list(node_list):
    if not node_list:
        return None
    head = ListNode(node_list[0])
    current = head
    
    for val in node_list[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

def main():
    head = build_list([1, 2, 2, 1])
    result = is_palindrome(head)
    
    print(result)

main()
````

### intesection of linked list


````python
class Node :
    def __init__(self,data):
        self.data = data
        self.next = None

def intersection_point(head1,head2):
    visNodes = set()
    
    temp1 = head1
    
    while temp1 :
        visNodes.add(temp1)
        temp1= temp1.next
    
    temp2 = head2
    
    while temp2:
        if temp2 in visNodes :
            return temp2
        
        temp2 =temp2.next
    return None

def main():
    head1 = Node(10)
    head1.next = Node(15)
    head1.next.next = Node(20)
    
    head2 = Node(5)
    head2.next = Node(10)
    
    #intersection point
    head2.next.next = head1.next
    
    itrpt = intersection_point(head1,head2)
    
    if itrpt :
        print(itrpt.data)
    else:
        print("None")

main()
    
    
````

Program for Nth node from the end of a Linked List

````python
class Node :
    def __init__(self,data):
        self.data = data
        self.next = None
def nthnodefromlast(head,n):
    #first travel n nodes from front with temp1
    # now temp1 will travel k-n nodes from now on and temp2 will travel k-n nodes i.e you are at nth npde from last
    
    temp1 = head
    temp2 = head
    for _ in range(n):
        temp1 = temp1.next
    
    while temp1 is not None:
        temp2 = temp2.next
        temp1=temp1.next
    
    return temp2.data

def main():
    # Create a hard-coded linked list:
    # 35 -> 15 -> 4 -> 20
    head = Node(35)
    head.next = Node(15)
    head.next.next = Node(4)
    head.next.next.next = Node(20)

    # Function Call to find the 4th node from end
    print(nthnodefromlast(head, 4))

main()
        
````
