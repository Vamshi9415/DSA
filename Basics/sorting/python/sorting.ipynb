```python
#selection sort select minimum and swap 

def selection_sort(arr):
    n = len(arr)
    
    for i in range(0,n-1):
        mini = i
        for j in range(i+1,n):
            if arr[j]< arr[mini]:
                mini = j
        
        arr[mini],arr[i] = arr[i], arr[mini]
    
    return arr

selection_sort([5,4,3,2,1])

# Time Complexity:
# Best Case: O(n²)
# Average Case: O(n²)
# Worst Case: O(n²)
# Space Complexity: O(1)
```

```python
# bubble sort push the element by adjacent swaps

def bubble_sort(arr):
    n = len(arr)
    didswap = False
    for i in range(n,0,-1):
        for j in range(0,i-1):
            if (arr[j]>arr[j+1]):
                arr[j],arr[j+1] = arr[j+1],arr[j]
                didswap = True
        
        if not didswap:
            print('entered did not swap ',i)
            break
    
    return arr

bubble_sort([1,2,3,4,5])
# bubble_sort([5,4,3,2,1])

# avg and wrost time complexity is o(n^2)
# best time complexity os o(n)
        
```

```python
#insertion sort -> takes an element and places in its correct oreder in sub array

def insertion_sort(a):
    n = len(a)
    
    for i in range(0,n):
        
        j = i
        
        while (j>0 and a[j-1]>a[j]):
            a[j-1],a[j] = a[j],a[j-1]
            j-=1
    
    return a

insertion_sort([5,4,3,2,1])

# best o(n) no swaps happen
# avg and worst (n^2)
```

```python
def merge(arr,low,mid,high):
    
    left = low
    right = mid+1
    temp = []
    while(left<mid+1 and right<=high):
        if(arr[left]<arr[right]):
            temp.append(arr[left])
            left+=1
        else:
            temp.append(arr[right])
            right+=1
    
    while(left<=mid):
        temp.append(arr[left])
        left+=1
    
    while(right<=high):
        temp.append(arr[right])
        right+=1
    
    
    for i in range(low,high+1):
        arr[i] = temp[i-low]
               
            
def mergesort(arr,low,high):
    if low >= high :
        return
    mid = (low + high) // 2
    mergesort(arr,low,mid)
    
    mergesort(arr,mid+1,high)
    
    merge(arr,low,mid,high)

arr= [5,4,3,2,1]
mergesort(arr,0,len(arr)-1)
print(*arr)
```

```python
def partition(arr, low, high):
    pivot = arr[low]
    left = low + 1
    right = high

    while True:
        while left <= right and arr[left] <= pivot:
            left += 1
        while left <= right and arr[right] > pivot:
            right -= 1
        if left > right:
            break
        arr[left], arr[right] = arr[right], arr[left]

    arr[low], arr[right] = arr[right], arr[low]
    return right

def quicksort(arr, low, high):
    if low < high:
        pivot_index = partition(arr, low, high)
        quicksort(arr, low, pivot_index - 1)
        quicksort(arr, pivot_index + 1, high)

arr = [5, 4, 3, 2, 1]
quicksort(arr, 0, len(arr) - 1)
print(*arr)

# Time Complexity:
# Best Case: O(n log n)   (when partition divides array evenly)
# Average Case: O(n log n)
# Worst Case: O(n^2)      (when partition is always most unbalanced, e.g., sorted array)
```
