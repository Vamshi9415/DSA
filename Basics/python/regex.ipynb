````python
import re

test = "my phone number is 123-456-7890" 

pattern = r'\d{3}-\d{3}-\d{4}'
match = re.search(pattern, test)
print(match)
print(match.group())
````

Here‚Äôs a **simple and clear** explanation of `find`, `findall`, and `search` in Python's `re` module:

---

## ‚úÖ `re.search()`

* **Finds the first match** in the *whole string*.
* Returns a **Match object** if found, otherwise `None`.

```python
import re
s = "abc123xyz456"

m = re.search(r'\d+', s)
print(m.group())
```

**Output:**

```
123
```

‚Üí Stops after the first match.

---

## ‚úÖ `re.findall()`

* Returns **all matches**.
* Returns a **list of strings** (all matches, without positions).

```python
import re
s = "abc123xyz456"

m = re.findall(r'\d+', s)
print(m)
```

**Output:**

```
['123', '456']
```

---

## ‚úÖ `re.finditer()`

* Returns **all matches**, but as **Match objects** (with index positions).
* Useful if you need **start/end positions**.

```python
import re
s = "abc123xyz456"

for m in re.finditer(r'\d+', s):
    print(m.group(), "at index", m.start())
```

**Output:**

```
123 at index 3
456 at index 9
```

---

## ‚úÖ `str.find()` (NOT regex)

* Searches without patterns ‚Üí **plain substring search**.
* Returns **index** of first occurrence, or `-1`.

```python
s = "hello world"
print(s.find("world"))
```

Output:

```
6
```

---

## üî• Summary Table

| Function        | Uses Regex? | Returns                  | Finds Multiple Matches? | Gives Index? |
| --------------- | ----------- | ------------------------ | ----------------------- | ------------ |
| `re.search()`   | ‚úÖ Yes       | 1 Match object           | ‚ùå No                    | ‚úÖ Yes        |
| `re.findall()`  | ‚úÖ Yes       | List of strings          | ‚úÖ Yes                   | ‚ùå No         |
| `re.finditer()` | ‚úÖ Yes       | Match objects (iterator) | ‚úÖ Yes                   | ‚úÖ Yes        |
| `str.find()`    | ‚ùå No        | Index (int)              | ‚ùå No                    | ‚úÖ Yes        |



````python
import re

# {n} - exactly n
text = "My number is 123-456-7890"
print(re.findall(r'\d{3}', text)) 

# {n,} - n or more
print(re.findall(r'\d{4,}', text))  

# {n,m} - between n and m
print(re.findall(r'\d{2,3}', text)) 

# Phone number pattern
phone = "123-456-7890"
print(re.search(r'\d{3}-\d{3}-\d{4}', phone))  

# Greedy vs Non-greedy
html = "<div>Hello</div><div>World</div>"
print(re.findall(r'<div>.*</div>', html)) 
print(re.findall(r'<div>.*?</div>', html)) 
````

````python
import re
text = "John Doe, Jane Smith"

pattern = r'(\w+) (\w+)'

matches = re.finditer(pattern,text)

for match in matches:
    print(match.group())
    print(match.group(1))
    print(match.group(2))
````

````python
text = "Email: test@gmail.com, admin@site.org"
pattern = r'(\w+)@(\w+)\.(\w+)'

for match in re.finditer(pattern, text):
    print(f"Full: {match.group()}")
    print(f"User: {match.group(1)}")
    print(f"Domain: {match.group(2)}")
    print(f"TLD: {match.group(3)}")
    print()
````
