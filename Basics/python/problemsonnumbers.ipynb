Check if a number is prime or not 

```python
import math
def is_prime(n):
    
    if n<2 :
        return False
    
    for i in range(2,int(math.sqrt(n)+1)):
        if n%i == 0:
            return False
    return True

is_prime(2)
```

Prime numbers in a given range

```python
for i in range(0, 100):
    if is_prime(i):
        print(i,end = " ")
```

```python
def is_Armstrong(n):
    num_str = str(n)
    num_digits = len(num_str)
    temp = n
    total = 0
    while temp>0:
        remain = temp%10
        
        total += remain**num_digits
        
        temp = temp //10
    
    return total == n

# is_Armstrong(123)

# we can also write 
def is_Armstrong2(n):
    num_str = str(n)
    num_digits = len(num_str)
    total = sum(int(digit)**num_digits for digit in num_str)
    
    return total == n

is_Armstrong2(153)
        
```

Check if a number is perfect number

a perfect number is a positive num that equals to sum of its proper divisors

```python
def is_perfect(n):
    if n<2:
        return False
    import math
    limit  = int(math.sqrt(n))+1
    
    sum =1
    for i in range(2,limit):
        if n%i == 0:
            # print(i)
            sum = sum+ i
            # if 4 is divisior of 28 then 28//4 is also a divisor of 28
            if i !=n//i:
                # print(n//i)
                sum = sum + n//i
            
            
    # print(sum)
    return sum  == n
is_perfect(28)
```

```python
# Maximum and Minimum digit in a number

def maxandmin(n):
    nstr = str(n)
    
    digits = [int(n) for n in nstr]
    
    return max(digits),min(digits)

maxandmin(9876544)
```

```python
# Print Fibonacci up to Nth Term 
#this recurssive fucntion can be used for finding single but for finding n terms its exponential time
def fib(n):
    if n == 0:
        return 0
    if n == 1 :
        return 1
    return fib(n-1)+fib(n-2)

def fibo(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=", ")
        a, b = b, a + b

fibo(10)

fibo(10)
```

```python
# gcd of two numbers 
#gcs also acalled hcf
def gcd(a,b):
    if b ==0:
        return a
    return gcd(b,b%a)

#GCD(a,b)×LCM(a,b)=a×b

def lcm(a,b):
    return (a*b)//gcd(a,b)
```

```python
# manual coding

def union_list(a,b):
    union = a.copy()
    
    for x in b :
        if x not in union:
            union.append(x)
    return union

def intersection(a,b):
    inter = []
    
    for x in a:
        if x in b and  x not in inter :
            inter.append(x)
    
    return inter

def difference(a,b):
    diff = []
    
    for x in a:
        if x not in b :
            diff.append(x)
    
    return diff

def symmetric_difference(a,b):
    return difference(a,b) +difference(b,a)

# Test cases for set operations

a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

print("Union:", union_list(a, b))              # [1, 2, 3, 4, 5, 6]
print("Intersection:", intersection(a, b))     # [3, 4]
print("Difference (a-b):", difference(a, b))   # [1, 2]
print("Difference (b-a):", difference(b, a))   # [5, 6]
print("Symmetric Difference:", symmetric_difference(a, b))  # [1, 2, 5, 6]

            
```

```python
# Permutations: Number of ways N people can occupy R seats (NPR)
# Formula: nPr = n! / (n - r)!

import math

def npr(n, r):
    if r > n or n < 0 or r < 0:
        return 0
    return math.factorial(n) // math.factorial(n - r)

# Example usage:
print(npr(5, 3))  # Output: 60
```

```python
#Replace all 0s with 1s in a given integer

def replace_1_with_0s(n):
    place = 0
    
    temp = n
    total = 0
    while temp >0 :
        digit = temp%10
        
        if digit == 0:
            digit = 1
        
        total = digit * (10**place) + total
        place+=1
        temp = temp//10
    
    print(total)

replace_1_with_0s(1001)
        
    
```
