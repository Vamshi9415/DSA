Memoization :- starts with the original problem and recursively breaks it into smaller sub problems.
Top Down Approach

```python
def fib(n,dp:list):
    if n<=1 :
        return n
    
    #if already computed then return stored value
    if dp[n]!= -1 :
        return dp[n]
    
    # other wise compute
    
    dp[n] = fib(n-1,dp) + fib(n-2,dp)
    
    return dp[n]

n=10
dp = [-1]*(n+1)
fib(n,dp)
    
    
```

Tabulation is a bottom up approach where we start from the base case and reach the final answer that we want
here we use iterative loops instead of loops

```python
def tfib(n):
    if n<=1 :
        return n
    
    dp = [0]*(n+1)
    
    dp[0] = 0
    dp[1] = 1
    
    for i in range(2,n+1):
        dp[i]= dp[i-1]+dp[i-2]
    
    return dp[n]

tfib(10)
```
