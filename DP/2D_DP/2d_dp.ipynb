Ninja is planing this •N' days-long training schedule. Each day. he can perform any one of these
three activities. (Running, ighting Practice or Learning New Moves). Each activity has some merit
points on each day. As Nit a has to improve all his skills. he can't do the same activity in two
consecutive days. Can you help Ninja find out the maximum merit points Ninja can earn?
You are given a 20 array of size •POINTS' with the points corresponding to each day and
activity. Your task is to calculate the maximum number of merit points that Ninja can earn.


```python
# Example 1:
# Input: matrix = [[10, 40, 70], [20, 50, 80], [30, 60, 90]]
# Output: 210
# Explanation:
# Day 1: fighting practice = 70
# Day 2: stealth training = 50
# Day 3: fighting practice = 90
# Total = 70 + 50 + 90 = 210
# This gives the optimal points.

# Example 2:
# Input: matrix = [[70, 40, 10], [180, 20, 5], [200, 60, 30]]
# Output: 290 
# Explanation:
# Day 1: running = 70
# Day 2: stealth training = 20
# Day 3: running = 200
# Total = 70 + 20 + 200 = 290
# This gives the optimal points.
```

```python
# pure recursion
def f(day,task,last):
    if day == 0:
        maxi = 0
        for i in range(0,3):
            if i != last :
                maxi = max(maxi,task[0][i])
        
        return maxi
    maxi = 0
    for i in range(0,3):
        
        if i != last :
            points = task[day][i] + f(day-1,task,i)
            maxi = max(maxi,points)
    
    return maxi

matrix_1 = [[10, 40, 70], [20, 50, 80], [30, 60, 90]]
matrix_2 = [[70, 40, 10], [180, 20, 5], [200, 60, 30]]

print(f(len(matrix_1)-1,matrix_1,3))

print(f(len(matrix_2)-1,matrix_2,3))
```

```python
# applying the memoization
def ninja_memo(day, task, last, dp):
    if dp[day][last] != -1:
        return dp[day][last]
    if day == 0:
        maxi = 0
        for i in range(3):
            if i != last:
                maxi = max(maxi, task[0][i])
        dp[0][last] = maxi
        return dp[0][last]

    maxi = 0
    for i in range(3):
        if i != last:
            points = task[day][i] + ninja_memo(day - 1, task, i, dp)
            maxi = max(maxi, points)
    dp[day][last] = maxi
    return dp[day][last]

matrix_1 = [[10, 40, 70], [20, 50, 80], [30, 60, 90]]
matrix_2 = [[70, 40, 10], [180, 20, 5], [200, 60, 30]]

dp1 = [[-1 for _ in range(4)] for _ in range(len(matrix_1))]
dp2 = [[-1 for _ in range(4)] for _ in range(len(matrix_2))]

print(ninja_memo(len(matrix_1) - 1, matrix_1, 3, dp1))
print(ninja_memo(len(matrix_2) - 1, matrix_2, 3, dp2))
```

Given two integers m and n, representing the number of rows and columns of a 2d array named matrix. Return the number of unique ways to go from the top-left cell (matrix[0][0]) to the bottom-right cell (matrix[m-1][n-1]).

Movement is allowed only in two directions from a cell: right and bottom.

```python
# Example 1:
# Input:
#  m = 3, n = 2  
# Output:
#  3  
# Explanation:
#  There are 3 unique ways to go from the top-left to the bottom-right cell:
# 1) right → down → down  
# 2) down → right → down  
# 3) down → down → right  

# Example 2:
# Input:
#  m = 2, n = 4  
# Output:
#  4  
# Explanation:
#  There are 4 unique ways to go from the top-left to the bottom-right cell:
# 1) down → right → right → right  
# 2) right → down → right → right  
# 3) right → right → down → right  
# 4) right → right → right → down
```

```python
def uniquePaths( m: int, n: int) -> int:
    def f(i, j):
        if i < 0 or j < 0:
            return 0
        if i == 0 and j == 0:
            return 1
        return f(i-1, j) + f(i, j-1)
    return f(m-1, n-1)

uniquePaths( 3, 3)
```

```python
# doing memoization

def uniquePathsMemo(m,n,dp):
    if dp[m][n] != -1:
        return dp[m][n]
    if m ==0 and n == 0:
        dp[0][0] = 1
        return dp[0][0]
    if m<0 or n<0 :
        return 0
    
    left = uniquePathsMemo(m,n-1,dp)
    up = uniquePathsMemo(m-1,n,dp)
    
    dp[m][n] = left + up
    
    return dp[m][n]

m = 2
n =4
dp = [[-1 for _ in range(n)] for _ in range(m)]

uniquePathsMemo(m-1,n-1,dp)
```

```python
# # applying tabulation
# tabulation is a bottom up way beacouse it starts from base case and accuamlate and accumulate
# steps to convert memoization to tabulation
# !. declare a base case and expresss all states in the loop and copy the recuurence and write

def uniquePathTabulation(m,n):
    dp = [[-1 for _ in range(n)] for _ in range(m)]
    
    for i in range(m):
        for j in range(n):
            if i == 0 and j ==0 :
                dp[0][0] = 1
                continue
            up = 0
            left = 0
            if i>0:
                up = dp[i-1][j]
            if j >0:
                left = dp[i][j-1]
            
            dp[i][j] = up +left
    return dp[m-1][n-1]


uniquePathTabulation(2,4)     
    
```

Grid Unique Paths 2 (DP 9)

Problem Statement: Given an m x n 2d array named matrix, where each cell is either 0 or 1. Return the number of unique ways to go from the top-left cell (matrix[0][0]) to the bottom-right cell (matrix[m-1][n-1]). A cell is blocked if its value is 1, and no path is possible through that cell.

Movement is allowed in only two directions from a cell - right and bottom.

```python
# Example 1:
# Input:
#  matrix = [[0, 0, 0], [0, 1, 0], [0, 0, 0]]  
# Output:
#  2  
# Explanation:
  
# The two possible paths to reach the bottom-right cell are:  
# 1) down → down → right → right  
# 2) right → right → down → down  
# Thus, the number of paths is 2.

# Example 2:
# Input:
#  matrix = [[0, 0, 0], [0, 0, 1], [0, 1, 0]]  
# Output:
#  0  
# Explanation:
#  There is no valid path to reach the bottom-right cell due to blocked cells (represented by 1).
```

```python

def f(i,j,grid,dp):
    if i<0 or j< 0 :
        return 0
    if grid[i][j] == 1:
        dp[i][j] = 0
        return dp[i][j]
    if dp[i][j] !=-1:
        return dp[i][j]
    
    if i == 0 and j == 0:
        dp[i][j] = 1
        return dp[i][j]
    
    
    dp[i][j] = f(i-1,j,grid,dp) + f(i,j-1,grid,dp)
    
    
    
    return dp[i][j]
def uniquePathsWithObstacles(obstacleGrid) -> int:
    m = len(obstacleGrid)
    n = len(obstacleGrid[0])
    dp = [[-1 for _ in range(n)] for _ in range(m)]
    return f(m-1,n-1,obstacleGrid,dp)

matrix = [[0, 0, 0], [0, 0, 1], [0, 1, 0]]  
uniquePathsWithObstacles(matrix)
        
```

Minimum Path Sum In a Grid (DP 10)

Problem Statement: Given a m x n grid filled with non-negative numbers, find a path from top left to bottom right, which minimizes the sum of all numbers along its path.
Note: You can only move either down or right at any point in time.

```python
# Input: grid = [[5,9,6],[11,5,2]]
# Output: 21

def f(i,j,grid):
        if i ==0 and j == 0 :
            return grid[0][0]
        if i<0 or j<0 :
            return float('inf')
        
        up = grid[i][j] + f(i-1,j,grid)
        left = grid[i][j] + f(i,j-1,grid)

        return min(up,left)
def minPathSum( grid ) -> int:
        return f(len(grid)-1,len(grid[0])-1,grid)
grid = [[5,9,6],[11,5,2]]
minPathSum(grid)
```

```python
#applying memeoization 
def f(i,j,grid,dp):
    if i ==0 and j == 0 :
        dp[i][j] = grid[0][0]
        return dp[i][j]
    
    if i<0 or j<0 :
        return float('inf')
    
    if dp[i][j] !=-1:
        return dp[i][j]
    
    up = grid[i][j] + f(i-1,j,grid,dp)
    left = grid[i][j] + f(i,j-1,grid,dp)
    dp[i][j] = min(up,left)
    return dp[i][j]

def minPathSum( grid) -> int:
    m = len(grid)
    n= len(grid[0])
    dp = [[-1 for _ in range(n)]for _ in range(m)]
    return f(m-1,n-1,grid,dp)

grid = [[5,9,6],[11,5,2]]
minPathSum(grid)
```
