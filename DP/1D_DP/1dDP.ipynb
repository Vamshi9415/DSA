You have been given a number of stairs. Initially, you are at the 0th stair, and you need to reach the Nth
stair. Each time you can either climb one step or two steps. You are supposed to return the number of
distinct ways in which you can climb from the 0th step to Nth step.

```python
#when they asked about count the num of ways or find min way or max way apply recursion then memoization and tabulation

#try to represent with an index, do all possible stuufs on that index based on problem statement

# sum of all stuffs -> count all ways
# min or max -> for min and max probs

def f(ind):
    if ind == 1:
        return 1
    if ind == 0 :
        return 1
    
    left = f(ind-1)
    right = f(ind-2)
    
    return left +right

f(3)
```

```python
# now apply memoization

def f(ind,dp):
    if ind<=1:
        return 1
    if dp[ind]!=-1 :
        return dp[ind]
    left = f(ind-1,dp)
    right = f(ind-2,dp)
    
    dp[ind] = left+right
    
    return dp[ind]

ind = 5
dp = [-1]*(ind+1)

print(f(ind,dp))
    
```

```python
#now apply tabulation

def f(ind):
    if ind<=1 :
        return 1
    
    dp = [-1]*(ind+1)
    
    dp[0] = 1
    dp[1] = 1
    
    for i in range(2,ind+1):
        dp[i] = dp[i-1] +dp[i-2]
    
    return dp[ind]

f(5)
    
    
```

Problem Statement: Given a number of stairs and a frog, the frog wants to climb from the 0th stair to the (N-1)th stair. At a time the frog can climb either one or two steps. A height[N] array is also given. Whenever the frog jumps from a stair i to stair j, the energy consumed in the jump is abs(height[i]- height[j]), where abs() means the absolute difference. We need to return the minimum energy that can be used by the frog to jump from stair 0 to stair N-1..

```python
# Example 1:
# Input: heights = [2, 1, 3, 5, 4]
# Output: 2
# Explanation: One possible route can be,
# 0th step -> 2nd Step = abs(2 - 3) = 1
# 2nd step -> 4th step = abs(3 - 4) = 1
# Total = 1 + 1 = 2.

# Example 2:
# Input: heights = [7, 5, 1, 2, 6]
# Output: 9
# Explanation: One possible route can be,
# 0th step -> 1st Step = abs(7 - 5) = 2
# 1st step -> 3rd step = abs(5 - 2) = 3
# 3rd step -> 4th step = abs(2 - 6) = 4
# Total = 2 + 3 + 4 = 9.

def f(index,heights):
    if index == 0 : return 0
    
    left = f(index-1,heights) + abs(heights[index] - heights[index-1])
    right = -1
    if index>1 :
        right = f(index-2,heights) + abs(heights[index] - heights[index-2])
    
    return min(left,right)


heights = [2, 1, 3, 5, 4]
n = len(heights) 

f(n-1,heights)



    
        
```

```python
#apply memoization
def f(index,heights,dp):
    if index == 0 : return 0
    if dp[index]!=-1 : return dp[index]
    left = f(index-1,heights,dp) + abs(heights[index] - heights[index-1])
    right = -1
    if index>1 :
        right = f(index-2,heights,dp) + abs(heights[index] - heights[index-2])
    
    dp[index] = min(left,right)
    return dp[index]


heights = [2, 1, 3, 5, 4]
n = len(heights) 
dp = [-1]*(n+1)
f(n-1,heights,dp)
```

```python
#applying tabulation

def frog_tabulation(heights):
    if not heights:
        return 0
    n = len(heights)
    dp = [-1]*(n+1)
    
    dp[0] = 0
    for i in range(1,n):
        jump_one = dp[i-1]+abs(heights[i]-heights[i-1])
        jump_two = -1
        if i>1:
            jump_two = dp[i-2]+abs(heights[i]-heights[i-2])
        
        dp[i] = min(jump_one,jump_two)
    
    return dp[n-1] 
            
heights = [2, 1, 3, 5, 4]

frog_tabulation(heights)
```

Problem Statement:

A frog wants to climb a staircase with n steps. Given an integer array heights, where heights[i] contains the height of the ith step, and an integer k. To jump from the ith step to the jth step, the frog requires abs(heights[i] - heights[j]) energy, where abs() denotes the absolute difference. The frog can jump from the ith step to any step in the range [i + 1, i + k], provided it exists. Return the minimum amount of energy required by the frog to go from the 0th step to the (n-1)th step.

```python
def frog_with_k_steps(a,k):
    n = len(a)
    dp = [-1]*(n+1)
    dp[0] = 0
    
    for i in range(1,n):
        min_cost = float('inf')
        
        for j in range(1,k+1):
            if i-j>=0:
                jump = dp[i-j]+abs(a[i]-a[i-j])
                
                min_cost = min(jump,min_cost)
        
        dp[i] = min_cost
    
    return dp[n-1]

height = [30, 10, 60, 10, 60, 50]
k = 2       

frog_with_k_steps(height,k)
            
    
    
```

Problem Statement: Given an array of N positive integers, we need to return the maximum sum of the subsequence such that no two elements of the subsequence are adjacent elements in the array.

```python
# Input: nums = [1, 2, 4]
# Output: 5
# Explanation: 
# Subsequence {1,4} gives maximum sum.

# Input:  [2, 1, 4, 9]
# Output: 11
# Explanation: 
# Subsequence {2,9} gives maximum sum

def nonadjsum(a,i,dp):
    if i == 0:
        return a[0]
    if i<0:
        return 0
    if dp[i]!=-1:
        return dp[i]
    
    pick = a[i]+nonadjsum(a,i-2,dp) #picking a[i] then not pick i-1 which is adjacent
    not_pick = 0+nonadjsum(a,i-1,dp)
    #here dp[x] = maximum sum in the subsequence with no adjacent sum in range[0..x]
    dp[i] = max(pick,not_pick)
    
    return dp[i]

def maximumnonadjacentsum(a):
    dp = [-1]*(len(a)+1)
    n = len(a)
    return nonadjsum(a,n-1,dp)

maximumnonadjacentsum([2,1,4,9])
```
